# seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips =sns.load_dataset('tips')
```

```python
tips.head()
```

# Univariate "one variable" data visualization

```python
# grouping with stripplot
sns.stripplot(x='sex', y='tip',hue='day', data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
# Strip plot
sns.stripplot(x='sex',hue='day', data=tips,jitter=True)
plt.title('Strip plot')
plt.xlabel('Gender')
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=tips, size=4,jitter=False)
plt.ylabel('tip($)')
plt.show()
```

```python
#Swarm plot
plt.figure(figsize=(10,5))
sns.swarmplot(x='day', y='tip', data=tips,facecolor='black')
# sns.violinplot(x='day',y='tip',data=tips.color='green')
plt.ylabel('tip($)')
plt.show()
```

```python
#Swarm plot
plt.figure(figsize=(10,5))
sns.swarmplot(x='day', y='tip',data=tips,facecolor='black')
sns.violinplot(x='day',y='tip',data=tips,color='black')
plt.ylabel('tip($)')
plt.show()
```

```python
# Density plot
sns.jointplot(x='total_bill', y='tip', data=tips,kind='kde')
plt.show()
```

# Box and violin plot

```python
#plt.subplot(1,2,1)
plt.figure(figsize=(10,10))
sns.boxplot(x='day', y='tip',data=tips,color='lightgrey')

plt.ylabel('tip($)')
plt.show()
```

```python
#plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')

plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip', data=tips)
plt.ylabel('tip($)')

plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')

plt.show()
```
